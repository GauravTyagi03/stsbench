```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os.path import join
from utils import loadmat
import os
import pandas as pd
import time

## POSTPROCESS EGOVIDEO DATA FOR EACH SESSION
## helper functions
def find_exp(date, task):
    exps = []
    for exp in range(10):
        ptb_path = f"../../fixation_{task}/{date}/tg_20{date[-2:] + date[0:-2]}_{exp}_data_structure.mat"
        if os.path.exists(ptb_path):
            exps.append(exp)
    return exps

dates = ["082824", "082924", "083024", "012925", "013025", "013125", "020425", "021125"] 
monkey_ids = ["T", "T", "T", "A", "A", "A", "A", "A"]
video_dims = [640, 360] # video dimensions in pixels are 640 x 320, width x height

all_neuron_ids = []
all_session_ids = []
all_monkey_ids = []

all_pxperdeg = []
all_fix_coords = []

for date, monkey_id in zip(dates, monkey_ids):
    st = time.time()
    exps = find_exp(date, 'egovideo')
    exp = exps[0]
    
    print(f"processing {date}...")
    # define paths to neural data, psychtoolbox data, and image_ids presented in each trial
    neural_path = f"../../../neural_data/{date}.pickle"

    ptb_path = f"../../fixation_egovideo/{date}/tg_20{date[-2:] + date[0:-2]}_{exp}_data_structure.mat"
    img_path = f"../../fixation_egovideo/{date}/video_ids.mat"
    save_dir = f"../../fixation_egovideo/neurips/preprocessed/"

    # open neural data
    with open(neural_path, "rb") as input_file:
        neural_data = pickle.load(input_file)

    # print some info about neural data and sync signals
    print("neural data info:")
    print(f"keys: {neural_data.keys()}")
    print(f"probe 0 keys: {neural_data[0].keys()}")
    print(f"number of single units: {len(np.unique(neural_data[0]['ids']))}")
    print(f"number of sync signals: {len((neural_data['sync_off']))}")
    neu_sync = neural_data['sync_off']
    
    print("ptb data info:")
    ptb_data = loadmat(ptb_path)
    print(f"keys: {ptb_data.keys()}")
    print(f"expsetup keys: {ptb_data['expsetup'].keys()}")
    print(f"number of sync signals: {len(ptb_data['expsetup']['stim']['edata_first_display'])}")
    ptb_sync = ptb_data['expsetup']['stim']['edata_loop_over']
    
    print("image data info:")
    img_data = loadmat(img_path)
    print(f"keys: {img_data.keys()}")
    print(f"number of im signals: {len(img_data['image_ids'])}")
    image_ids = img_data['image_ids']
    
    # determine what index to align the neural and ptb data to 
    neu_sync_a = neu_sync - neu_sync[0]
    ptb_sync_a = ptb_sync - ptb_sync[0]

    dist = np.zeros(len(neu_sync_a) - len(ptb_sync_a) + 1)
    for i in range(len(neu_sync_a)):
        if i + len(ptb_sync_a) <= len(neu_sync_a):
            dist[i] = np.sqrt(np.sum(np.square(neu_sync_a[i:i+len(ptb_sync_a)] - neu_sync_a[i] - ptb_sync_a)))

    match_idx = np.argmin(dist)

    print(f"neural data is aligned to the {match_idx + 1}st ptb sync signal")

    
    et = time.time()
    print('time for loading ' + str(et-st))

    st = time.time()

    # plot the alignment to be sure
    i = match_idx
    plt.plot(np.square(neu_sync_a[i:i+len(ptb_sync_a)] - neu_sync_a[i] - ptb_sync_a))
    plt.xlabel("trial")
    plt.ylabel("timing error (sec)")
    plt.title(date)
    plt.show()
    
    assert min(dist) < .5 # assert that the alignment is good 

    # get probe onset times relative to start of the trial
    n_trials = len(ptb_sync_a)
    n_probes_per_trial = ptb_data['expsetup']['stim']['probe_number']
    probe_onsets = np.zeros((n_trials, n_probes_per_trial))
    for trial_id in range((n_trials)):
        trial_start = ptb_data['expsetup']['stim']['edata_first_display'][trial_id]
        trial_end = ptb_data['expsetup']['stim']['edata_loop_over'][trial_id]
        probe_on = ptb_data['expsetup']['stim']['edata_probe_on'][trial_id]
        onsets = probe_on - trial_start
        probe_onsets[trial_id, :onsets.size] = onsets
    
    # label trials as train trials or test trials based on whether images were shown repeatedly
    all_image_ids = []
    for i in range(len(image_ids)):
        all_image_ids = all_image_ids + (list(image_ids[i][:]))

    uq, freq = np.unique(all_image_ids, return_counts=True)
    res = {uq[i]: freq[i] for i in range(len(uq))}

    train_trials = []
    for i in range(len(image_ids)):
        all_cnts = []
        for probe_id in range(len(image_ids[i])):
            all_cnts.append(res[image_ids[i][probe_id]])
        all_cnts = np.array(all_cnts)

        if np.any(all_cnts > 1):
            train_trials.append(0)
        else:
            train_trials.append(1)

    et = time.time()
    print('time for train/test ' + str(et-st))
    
    st = time.time()
            
    # get neural data 0.06 sec before probe onset and 0.14 sec after probe offset for each probe
    pre_period = 0.06
    post_period = 0.14 
    probe_duration = 0.2
    n_times = int((probe_duration + pre_period + post_period) * 1000)
    n_neurons = np.max(neural_data[0]["ids"]) + 1
    times = np.linspace(-pre_period, probe_duration + post_period, n_times)
    
    # get condition label    
    train_label = []
    image_label = []
    spike_trains = []
    probe_order = []
        
    # efficient way for searching for spikes    
    trial_starts = np.concatenate((neural_data['sync_on'][match_idx:], [neural_data['sync_on'][match_idx:][-1] + 10]))
    start_idxes = np.searchsorted(neural_data[0]["times"], trial_starts)

        
    et = time.time()
    print('time for search sorting ' + str(et-st))
    st = time.time()

    for trial_id in range((n_trials)):
        # if correct, use all probes
        process_trial = False
        if ptb_data['expsetup']['stim']['edata_trial_online_counter'][trial_id]:
            n_probes_total = n_probes_per_trial
            process_trial = True
        # else use all but last probe shown
        elif type(ptb_data['expsetup']['stim']['edata_probe_on'][trial_id]) != float and len(ptb_data['expsetup']['stim']['edata_probe_on'][trial_id]) > 1:
            n_probes_total = len(ptb_data['expsetup']['stim']['edata_probe_on'][trial_id]) - 1
            process_trial = True

        # process trial
        if process_trial:
            trial_neural_times = neural_data[0]["times"][start_idxes[trial_id]:start_idxes[trial_id+1]]
            trial_neural_ids = neural_data[0]["ids"][start_idxes[trial_id]:start_idxes[trial_id+1]]

            for probe_id in range(n_probes_total):
                # get spikes in trial X aligned to the start of the trial
                probe_start = neural_data['sync_on'][trial_id+match_idx] + probe_onsets[trial_id, probe_id]

                include_idx = np.logical_and(trial_neural_times > probe_start - pre_period, trial_neural_times < probe_start + probe_duration + post_period)
                probe_times = trial_neural_times[include_idx] - probe_start
                probe_ids = trial_neural_ids[include_idx]
                
                spike_train = np.zeros((n_times, n_neurons), dtype=bool)
                for neuron_id in range(n_neurons):
                    indices = (np.floor(probe_times[probe_ids == neuron_id] * 1000) + pre_period * 1000).astype(int)
                    spike_train[indices, neuron_id] = 1

                image_label.append(image_ids[trial_id][probe_id])
                train_label.append(train_trials[trial_id])
                spike_trains.append(spike_train)
                probe_order.append(probe_id)
    
    spike_trains = np.array(spike_trains)
    
    et = time.time()
    print('time for fetching neural data ' + str(et-st))
    
    # saving output - spike trains
    for neuron_id in range(n_neurons):
        all_neuron_ids.append(neuron_id)
        all_session_ids.append(date)
        all_monkey_ids.append(monkey_id)
        
        screenpx_to_vidpx = (ptb_data['expsetup']['screen']['dispcenter']*2)[0]/video_dims[0]
        assert screenpx_to_vidpx == 1920/640 # this should always be 1920/640
        
        all_pxperdeg.append(ptb_data['expsetup']['screen']['deg2pix']/screenpx_to_vidpx) # this is now in video px per deg
        all_fix_coords.append(ptb_data['expsetup']['stim']['esetup_fixation_coord'][0]) # this is in deg visual angle
    
    np.save(join(save_dir, f"{date}_video.npy"), spike_trains)
    
    # save session_df
    df = pd.DataFrame({"train": train_label, "image": image_label, "order": probe_order})
    df.to_csv(join(save_dir, f"{date}_video.csv"))
    
# save times
np.save(join(save_dir, f"times_video.npy"), times)

# save neuron_df
df = pd.DataFrame({
    'neuron_id': all_neuron_ids,
    'session_id': all_session_ids,
    'monkey_id': all_monkey_ids,
    'px_per_deg': all_pxperdeg, 
    'fix_coord_deg': all_fix_coords,
})

df.to_csv(join(save_dir, f"neuron_data_video.csv"))

print("all done!")
```
